```python
age_groups = [(23,30), (31,40), (41,50), (51,60), (61,70),(71,85)]

def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return("Unknown")
```

```python
food_df['age_group'] = food_df['Age'].apply(assign_age_group)
```

```python
food_df[['Age', 'age_group']]
```

```python
age_order = ['23 - 30', '31 - 40', '41 - 50', '51 - 60', '61 - 70','71 - 85']

sns.pointplot(data=food_df, x='age_group', y='accepted_campaigns', order=age_order)
```

```python
age_counts = food_df['age_group'].value_counts()
age_counts
```

```python
age_accepted_campaign = food_df[food_df['accepted_campaigns'] == 1]['age_group'].value_counts()
age_accepted_campaign
```

```python
percentage = age_accepted_campaign / age_counts
percentage
```

```python
percent_age_group = percentage.reset_index()
```

```python
percent_age_group.columns = ['age_group', 'percentage']
```

```python
percent_age_group = percent_age_group.sort_values('age_group')
percent_age_group
```

```python
# Shows the the percent of each age group that accepted any campaign
sns.barplot(x='age_group', y='percentage', data=percent_age_group)
```

# Overall Findings

## Age Demographics
- The age groups 41-70 is our largest customer base

## Purchase Area
- Most of our customers purchase either on the web or in the physical store.

## Marital Demographics
- Married is the biggest customer base for us

## Children Demographics
- People with no kids spent more and accepted campaigns more often so it would be good to continue to focus on that demographic.

## Education Demographics
- Graduation level education spend far more in our stores than any other education level.
- They are in the middle as far as currently accepting our campaigns

# Conclusion
- Our largest customer bases are accepting less than 30% of the marketing campaigns we put out.
- We should focus our campaigns on married couples between the ages of 41-70 who shop either in the store or online.
- We can also focus some campaigns on customers that have no children and have graduation level education.
- Currently our campaigns seem to be most effective to the age groups 23-30, 71-85, widows, and those with a PhD education level. However, even though they accept our campaigns at the highest rate they are generally spending significantly less money than their demographic counterparts meaning we are gaining very little for our current campaigns.
